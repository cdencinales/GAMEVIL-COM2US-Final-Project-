#what are the most popular free games in the top charts of google play.

```python
%%sql
SELECT title_name, title_developer, category_ID, title_rating, num_reviews, DENSE_RANK() OVER(ORDER BY title_rating DESC) my_rank
FROM free_games
GROUP BY num_reviews;
```

#what are the most popular paid games in the top charts of google play.

```python
%%sql
SELECT title_name, title_developer, category_ID, title_rating, num_reviews, title_price, DENSE_RANK() OVER(ORDER BY title_rating DESC) my_rank
FROM paid_games
GROUP BY num_reviews;
```

#through these querys (group by, order by, window function) we find what the highest rated games (by paid or free) and group them in order from most to least reviews in their particular rating bracket. additionally, through the denserank() function we get a rank assoicated to a game in every row to further help us visualize the data. 

#what are the most popular games both paid and free.

```python
%%sql
CREATE VIEW all_games AS
SELECT title_name, title_developer, category_ID, title_rating, num_reviews, 'FREE' AS Free_or_Paid
FROM free_games
UNION
SELECT title_name, title_developer, category_ID, title_rating, num_reviews, 'PAID'
FROM paid_games;

SELECT *
FROM all_games;
```

#the query above shows a union of both free and paid games stored in a view named 'all_games' this will allow us to analyze the most popular categories in both free and paid charts while keeping the data readable and easy to understand. Wether the game is free or paid is shown in the last column called Free_or_Paid. The issue with analyzing the charts in this manner is that paid and free games are fundamentally different. The development, in-game transactions and the general enviroment of the game will differ greatly if it is paid or free. 

#Now we will analyze what the most popular categories are in all_games.

```python
%%sql
SELECT category_ID, count(category_ID) as instances, DENSE_RANK() OVER (ORDER BY category_ID) my_rank
FROM all_games 
GROUP BY category_ID;
```

#this query allows us to see that the top categories for all games are action closely followed by simulation, puzzle and casual.

#Now we will see the top games in each of the above mentioned categories.

```python
%%sql
#Action
SELECT title_name, title_developer, title_rating, num_reviews,DENSE_RANK() OVER(ORDER BY num_reviews desc, title_rating DESC), Free_or_Paid
FROM all_games
WHERE category_ID = 1;
```

```python
%%sql
#Puzzle
SELECT title_name, title_developer, title_rating, num_reviews,DENSE_RANK() OVER(ORDER BY num_reviews desc, title_rating DESC), Free_or_Paid
FROM all_games
WHERE category_ID = 3;
```

#Now we will check if our company has any apps in the top charts. Since our company currently only has free apps available we will refrain from using the paid games table.

```python
%%sql
#gamevil
SELECT title_name, title_rating, num_reviews
FROM gamevil_games
WHERE (
		SELECT title_name
		FROM free_games
		WHERE title_developer = 'GAMEVIL');
```

```python
%%sql
#come2us
SELECT title_name, title_rating, num_reviews
FROM come2us_games
WHERE (
		SELECT title_name
		FROM free_games
		WHERE title_developer = 'COME2US');
```

#as we can see from the results of the subqueries above, there are no games from our companies currently in the top charts, this could be for a number of reasons including but not limited to the fact that we lack abundance of games in terms of the most popular categories in the top charts. 

#Now we will study our app rating and break them down into different group so we know what is working and what has to improve. For this we will first create a view of a union between the 2 companies so that all our apps are in one place.

```python
%%sql
CREATE VIEW company_games AS
SELECT title_name, title_developer, category_ID, title_rating, num_reviews
FROM gamevil_games
UNION
SELECT title_name, title_developer, category_ID, title_rating, num_reviews
from come2us_games;
```

```python
%%sql
SELECT *
FROM company_games;
```

#As we can see most of our games have great ratings, however there are some that could improve and one that has to improve.


```python
%%sql
SELECT title_name,
CASE
	WHEN title_rating >= 4 THEN 'Great Rating'
	WHEN title_rating BETWEEN 3 and 4 THEN 'Good Rating, Could Improve'
	ELSE 'Rating Too Low'
END
FROM company_games
ORDER BY title_rating DESC;
```
